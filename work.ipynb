## タスク学習の確認

```python
#ベース
import os
import regex as re
import matplotlib.pyplot as plt

ROOT_DIR = "/home/k-yanase/qnap5/k-yanase/Klab_MultiModalModel/task_train/"
```

```python
def get_loss_list(filepath:str,phase:str="Train"):
    loss_list = []
    pat = rf"\[Epoch \((\d+)/\d+\) {phase}\] Loss : ([0-9.]+)"
    with open(filepath):
        log = re.findall(pat,open(filepath).read())
    for epoch,loss in log:
        loss_list.append(float(loss))
    return loss_list
```

```python
#VQA
DATA_DIR = os.path.join(ROOT_DIR,"vqa")

train_loss = get_loss_list(os.path.join(DATA_DIR,"train.log"))
val_loss = get_loss_list(os.path.join(DATA_DIR,"val.log"),phase="Val")

plt.plot(val_loss,label="val_loss")
plt.plot(train_loss,label="train_loss")
plt.legend()
plt.show()
```
